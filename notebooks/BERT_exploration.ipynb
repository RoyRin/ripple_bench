```python
# Use a pipeline as a high-level helper
from transformers import pipeline
import torch

pipe = pipeline("fill-mask", model="google-bert/bert-base-uncased")


# Load model directly
from transformers import AutoTokenizer, AutoModelForMaskedLM

tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
model = AutoModelForMaskedLM.from_pretrained("google-bert/bert-base-uncased")
```

```python
text = "The capital of France is [MASK]."
print("Pipeline predictions:")
results = pipe(text)
for result in results:
    print(f"Token: {result['token_str']}, Score: {result['score']:.4f}")

# Directly loading the model and tokenizer
tokenizer = AutoTokenizer.from_pretrained("google-bert/bert-base-uncased")
model = AutoModelForMaskedLM.from_pretrained("google-bert/bert-base-uncased")

# Prepare the input text with a mask token
input_text = "The capital of France is [MASK]."
input_ids = tokenizer.encode(input_text, return_tensors="pt")
mask_token_index = torch.where(input_ids == tokenizer.mask_token_id)[1]

# Get the predictions from the model
with torch.no_grad():
    output = model(input_ids)
logits = output.logits

# Focus on the logits of the mask token
mask_token_logits = logits[0, mask_token_index, :]

# Get the top 5 predicted tokens
top_5_tokens = torch.topk(mask_token_logits, 5, dim=1).indices[0].tolist()

print("\nDirect model predictions:")
for token in top_5_tokens:
    word = tokenizer.decode([token]).strip()
    print(f"Predicted token: {word}")
```

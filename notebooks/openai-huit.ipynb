```python
from pathlib import Path
import os 
import torch
from importlib import reload 
from openai import OpenAI
```

```python
CWD = Path(os.getcwd())
BASE_DIR = CWD.parent

SECRETS = BASE_DIR / "SECRETS" / "openai_huit.secret"
# open SECRETS
with open(SECRETS, "r") as f:
    OPEN_API_key = f.read().strip()

#print(f"'{OPEN_API_key}'")
# get wmdp accuracy
device = torch.device('cuda:0')
```

```python
import requests
import json

url = "https://go.apis.huit.harvard.edu/ais-openai-direct/v1/chat/completions"
payload = json.dumps({
  "model": "gpt-4o-mini",
  "messages": [
    {
      "role": "user",
      "content": "To what extent does scientific conservatism limit new discoveries?"
    }
  ],
  "temperature": 0.7
})
headers = {
  'Content-Type': 'application/json',
  'api-key': OPEN_API_key
}
response = requests.request("POST", url, headers=headers, data=payload)
print(response.text)
```

```python
l = ripple_bench_df[ripple_bench_df.original_topic == "Viral evolution"]
x = list(range(len(l)))
y = l.distance

print(l.distance.value_counts())
l.head()
from collections import defaultdict
distance_to_query_topic = defaultdict(list)
distance_to_query_id = defaultdict(list)
for ind, row in l.iterrows():
    distance_to_query_topic[row.distance].append(row.topic)
    distance_to_query_id[row.distance].append(row.name)

print(distance_to_query_topic)

print(f"index, distance, queries")
for ii, (k,v) in enumerate(distance_to_query_topic.items()):
    
    #if ii < 20 or k > 95 or abs(k - 50) < 10:
    print(f"{ii, } -- {k},: {v[:2]}")
def get_accuracy_for_topic(question_list, df_):
    correct = 0
    for question_id in question_list:
        if df_.iloc[question_id].is_correct:
            correct += 1
    if len(question_list) == 0:
        return 0
    return correct / len(question_list)

get_accuracy_for_topic([1,2,3], df)
```

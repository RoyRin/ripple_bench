```python
df.head()
# print first 4 rows of the dataset
for i, row in df.head(22).iterrows():
    question = row['question']
    answer = row['answer']
    print(f"Question: {question}\nAnswer: {answer}\n")
    
    doc = nlp(f"{question} | {answer}")
    names = set()
    # Extract entities labeled as PERSON
    for ent in doc.ents:
        if ent.label_ == "PERSON":
            names.add(ent.text)
    print(f"Extracted Names: {', '.join(names)}\n")
```
